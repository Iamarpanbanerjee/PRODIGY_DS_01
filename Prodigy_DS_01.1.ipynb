```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'C:\Users\arpan\Downloads\Internship\Prodigy Infotech\PRODIGY_DS_01\WorldBank_Dataset\Metadata_Country_API_SP.POP.TOTL_DS2_en_csv_v2_85.csv')
```

```python
df
```

```python
# Extract relevant columns
data = df[['IncomeGroup', 'TableName']]

# Drop NaN values
data = data.dropna()

```

```python
data
```

```python
# Group by 'IncomeGroup' and count the occurrences of each income group
grouped_data = data.groupby('IncomeGroup').count()

# Plotting bar graph with counts
plt.figure(figsize=(10, 6))
bars = plt.bar(grouped_data.index, grouped_data['TableName'])
plt.xlabel('Income Group')
plt.ylabel('Count')
plt.title('Count of Countries by Income Group')
plt.xticks(rotation=45)
plt.grid(True)

# Display counts on top of each bar
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 2), ha='center', va='bottom')

plt.tight_layout()
plt.show()

# Plotting histogram
plt.figure(figsize=(10, 6))
plt.hist(data['IncomeGroup'], bins=len(grouped_data), edgecolor='black')
plt.xlabel('Income Group')
plt.ylabel('Frequency')
plt.title('Histogram of Income Groups')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Plotting the count plot for Region and IncomeGroup
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Region', hue='IncomeGroup')
plt.title('Count of Income Groups in Each Region')
plt.xlabel('Region')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Income Group')
plt.show()
```
